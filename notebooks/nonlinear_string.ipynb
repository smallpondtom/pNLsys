# Nonlinear String Model

![Diagram](../images/string_diagram.png)

## Derivation

a) First we must define variables and parameters which characterize the nonlinear string problem shown in the above figure.
- $l$: length of the string [$m$]
- $\rho$: density of string [$kg/m^3$]
- $A$: cross-sectional area of the string [$m^2$]
- $E$: Young's modulus of the string [$N/m^2$]
- $T_0$: pre-tension [$N$]
- $u$: longitudinal displacement [$m$]
- $v$: lateral displacement [$m$]
- $w$: vertical displacement [$m$]
The dynamic exerted on a small portion of the string is shown below.

This problem is has geometric nonlinearity and material nonlinearity. For the stress we add an additional pre-stress term as follows. 
\begin{gather*}
    \sigma = \dfrac{T_0}{A} + E_1\epsilon + E_2\epsilon^2.
\end{gather*}
To find the equation of motion we use the Hamilton's principle. Firstly, the kinetic energy is 
\begin{gather*}
    T = \dfrac{1}{2}\int_0^l{\rho A \left( \dot{u}^2 + \dot{v}^2 + \dot{w}^2 \right)}dx
\end{gather*}
Recall that for a 1-D string system the only non-zero strain term is $\epsilon_{11}$, and 
\begin{align*}
    \epsilon_{11} &= \dfrac{1}{2}\left( u_{i,j} + u_{j,i} + u_{k,i} \cdot u_{k,j} \right)\\
    &= \dfrac{1}{2}\left( u_{x} + u_{x} + u_{x} \cdot u_{x} + v_{x} \cdot v_{x} + w_{x} \cdot w_{x} \right)\\
    &= u_{x} + \dfrac{1}{2}\left( u_{x}^2 + v_{x}^2 + w_{x}^2 \right)
\end{align*}
The potential energy can be computed as follows.
\begin{align*}
    u &= \int{\sigma_xx}{\epsilon_{11}}\\
    &= \int{\left(\dfrac{T_0}{A} + E_1\epsilon_{11} + E_2\epsilon^2_{11} \right)}{\epsilon_{11}}\\
    &= \dfrac{T_0}{A}\epsilon_{11} + \dfrac{1}{2}E_1\epsilon_{11}^2 + \dfrac{1}{3}E_2\epsilon_{11}^3 
\end{align*}
Then 
\begin{align*}
    U &= \int_0^l{uA}dx\\
    &= \int_0^l{\left( T_0\epsilon_{11} + \dfrac{1}{2}E_1A\epsilon_{11}^2 + \dfrac{1}{3}E_2A\epsilon_{11}^3 \right)}dx
\end{align*}
The Lagrangian becomes 
\begin{align*}
    L = T - U = \dfrac{1}{2}\int_0^l{\rho A \left( \dot{u}^2 + \dot{v}^2 + \dot{w}^2 \right)}dx - \int_0^l{\left(T_0\epsilon_{11} + \dfrac{1}{2}E_1\epsilon_{11}^2 + \dfrac{1}{3}E_2\epsilon_{11}^3 \right)}dx.
\end{align*}
Thus,
\begin{align*}
    \delta L = \int_0^l{\big[ \rho A \left(  \dot{u}\delta \dot{u} + \dot{v}\delta\dot{v} + \dot{w}\delta\dot{w}\right) - \left(T_0\epsilon_{11} +  E_1 A\epsilon_{11} + E_2A\epsilon_{11}^2 \right)\delta\epsilon_{11} \big]}dx.
\end{align*}
Now since $\delta W^{NC} = 0$, the extended Hamilton's principle 
\begin{align*}
    \int_{t_1}^{t_2}{\big[ \delta L + \delta w^{NC} \big]}dt = 0
\end{align*}
becomes
\begin{gather*}
    \int_{t_1}^{t_2}{ \int_0^l{\big[ \rho A \left(  \dot{u}\delta \dot{u} + \dot{v}\delta\dot{v} + \dot{w}\delta\dot{w}\right) - \left(T_0\epsilon_{11} + E_1 A\epsilon_{11} + E_2A\epsilon_{11}^2 \right)\delta\epsilon_{11} \big]}dx}dt = 0
\end{gather*}
where 
\begin{gather*}
    \delta\epsilon_{11} = (1 + u_{x})\delta u_{x} + v_{x}\delta v_{x} + w_{x}\delta w_{x}.
\end{gather*}
Using integration by parts we have
\begin{align*}
    \int_0^l{\int_{t_1}^{t_2}{\left(\rho A \dot{u}\delta \dot{u} \right)}dt}dx &= \int_0^l{\rho A \Big\{ \big[ \dot{u}\delta u \big]^{t_2}_{t_1} - \int_{t_1}^{t_2}{\ddot{u}\delta u}dt \Big\}}dx\\
    &= -\int_0^l{\int_{t_1}^{t_2}{\Big( \rho A \ddot{u}\delta u \Big)}dt}dx \hspace{0.8cm}\because \delta u(t_1) = \delta u(t_2) = 0\\
\end{align*}
Similarly,
\begin{align*}
    \int_0^l{\int_{t_1}^{t_2}{\left(\rho A \dot{v}\delta \dot{v} \right)}dt}dx 
    &= -\int_0^l{\int_{t_1}^{t_2}{\Big( \rho A \ddot{v}\delta v \Big)}dt}dx \\
    \int_0^l{\int_{t_1}^{t_2}{\left(\rho A \dot{w}\delta \dot{w} \right)}dt}dx 
    &= -\int_0^l{\int_{t_1}^{t_2}{\Big( \rho A \ddot{w}\delta w \Big)}dt}dx 
\end{align*}
We also use integration by part for the following.
\begin{align*}
    & \int_{t_1}^{t_2}{\int_0^l{\sigma A \left( 1 + u_{x} \right) \delta u_{x}}dx}dt\\
    &= \int_{t_1}^{t_2}{\Bigg\{ \bigg[ \sigma A (1 + u_{x} ) \delta u \bigg]_0^l - \int_0^l{\Big( \sigma A (1 + u_{x} ) \Big)_{x} \delta u  }dx \Bigg\} }dt\\[0.5cm]
    & \phantom{--} \text{from the boundary condition }  u(l, t) = u(0, t) = 0\\[0.5cm]
    &= -\int_{t_1}^{t_2}{\int_0^l{\Big( \left(T_0 \epsilon_{11} + E_1 A\epsilon_{11} + E_2A\epsilon_{11}^2 \right) (1 + u_{x} ) \Big)_{x} \delta u  }dx}dt
\end{align*}
Similarly,
\begin{align*}
    \int_{t_1}^{t_2}{\int_0^l{\sigma A v_{x}\delta v_{x}}dx}dt &= -\int_{t_1}^{t_2}{\int_0^l{\left( \left(T_0\epsilon_{11} + E_1 A\epsilon_{11} + E_2A\epsilon_{11}^2 \right)v_{x} \right)_{x}\delta v}dx}dt\\[0.2cm]
    \int_{t_1}^{t_2}{\int_0^l{\sigma A w_{x}\delta w_{x}}dx}dt &= -\int_{t_1}^{t_2}{\int_0^l{\left( \left(T_0\epsilon_{11} + E_1 A\epsilon_{11} + E_2A\epsilon_{11}^2 \right)w_{x} \right)_{x}\delta w}dx}dt
\end{align*}
Finally by combining everything we get the following equation of motion 
\begin{align*}
    \rho A \ddot{u} - \left( \left(T_0\epsilon_{11} + E_1 A\epsilon_{11} + E_2A\epsilon_{11}^2 \right) (1 + u_{x} ) \right)_{x} = 0\\
    \rho A \ddot{v} - \left( \left(T_0\epsilon_{11} + E_1 A\epsilon_{11} + E_2A\epsilon_{11}^2 \right)v_{x} \right)_{x} = 0\\
    \rho A \ddot{w} - \left( \left(T_0\epsilon_{11} + E_1 A\epsilon_{11} + E_2A\epsilon_{11}^2 \right)w_{x} \right)_{x} = 0
\end{align*}

## Including Kelvin-Voigt Damping

if we include the Kelvin-Voigt damping 
\begin{gather*}
    \sigma = \dfrac{T_0}{A} + E_1\epsilon_{11} + E_2\epsilon_{11} + \alpha\dot{\epsilon_{11}}.
\end{gather*}
Hence, the derivation of the equation of motion is identical to the steps that we took for part (a), and thus, it becomes 
\begin{align*}
    \rho A \ddot{u} - \left( \left(T_0 + E_1 A\epsilon_{11} + E_2A\epsilon_{11}^2 + \alpha A \dot{\epsilon_{11}}\right) (1 + u_{x} ) \right)_{x} = 0\\
    \rho A \ddot{v} - \left( \left(T_0 + E_1 A\epsilon_{11} + E_2A\epsilon_{11}^2 + \alpha A \dot{\epsilon_{11}}\right)v_{x} \right)_{x} = 0\\
    \rho A \ddot{w} - \left( \left(T_0 + E_1 A\epsilon_{11} + E_2A\epsilon_{11}^2 + \alpha A \dot{\epsilon_{11}}\right)w_{x} \right)_{x} = 0
\end{align*}

## Galerkin's method of approximating PDE to ODE

### Linear analysis
(a) The given condition of $u(x, t) = w(x, t) = 0$ implies that 
\begin{gather*}
    u_{x} = u_{xx} = \dot{u} = \ddot{u} = w_{x} = w_{xx} = \dot{w} = \ddot{w} = 0. 
\end{gather*}
Additionally, if we assume that the tension is not effected by the displacement $v$, we can disregard the first equation and can reduces the equations to 
\begin{gather*}
    \rho A \ddot{v} - \Big( \left(T_0 + E_1 A\epsilon_{11} + E_2A\epsilon_{11}^2 + \alpha A \dot{\epsilon_{11}}\right)v_{x} \Big)_{x} = 0\\
    \epsilon_{11} = \dfrac{1}{2}v^2_{x} 
\end{gather*}
Now if we linearize the equation governing $v(x, t)$ while keeping the nonlinear terms associated with the damping term $\alpha$, we get 
\begin{gather*}
    \rho A \ddot{v} - (T_0 v_{x} + \alpha Av^2_{x}\dot{v}_{x})_{x} = 0
\end{gather*}
which then becomes the single PDE governing the free motions as follows.
\begin{gather*}
    \rho A \ddot{v} - T_0v_{xx} - 2\alpha A v_{x}v_{xx}\dot{v}_{x} - \alpha A v^2_{x}\dot{v}_{xx} = 0 .
\end{gather*}
The terms arising from the Kelvin-Voigt damping are colored in blue in the answer. The solution shows that with the nonlinear terms with damping, the linear system will have a contribution due to damping.


### Nonlinear PDE to Nonlinear ODE
(b) Going back to the system of 
\begin{gather*}
    \rho A \ddot{v} - \left( \left(T_0 + E_1 A\epsilon_{11} + E_2A\epsilon_{11}^2 + \alpha A \dot{\epsilon_{11}}\right)v_{x} \right)_{x} = 0\\
    \epsilon_{11} = \dfrac{1}{2}v^2_{x} + \lambda  
\end{gather*}
with the boundary conditions of $v(0, t) = v(l, t) = 0$ with the pinned-pinned string. From the linear system that we have derived in Problem 2.a. and from the boundary condition, we can deduce the basis equations (eigenfunctions) for the Galerikin method to be (with first and second mode)
\begin{gather*}
    \phi_1 = \sin \Big(\dfrac{\pi x}{l} \Big)\\
    \phi_2 = \sin\Big(\dfrac{2 \pi x}{l} \Big).
\end{gather*}
And, with the two-term approximation we have
\begin{gather*}
    \tilde{v}(x, t) = \beta_1(t)\phi_1(x) + \beta_2(t)\phi_2(x).
\end{gather*}
Let, 
\begin{gather*}
    \gamma(x, t) = T_0 + E_1 A\epsilon_{11} + E_2 A \epsilon^2_{xx} + \alpha A \dot{\epsilon}_{xx}.
\end{gather*}


Going back to our governing equation, we have (plug in $\tilde{v}$ and $\tilde{\gamma}$ into the governing equation)
\begin{align*}
    \rho A \ddot{v} - \left( \left(T_0 + E_1 A\epsilon_{11} + E_2A\epsilon_{11}^2 + \alpha A \dot{\epsilon_{11}}\right)v_{x} \right)_{x} &= 0\\
    \rho A \ddot{v} - \gamma_x v_x - \gamma v_{xx} &= 0
\end{align*}
Then 
\begin{align*}
    \tilde{\gamma} &= T_0 + \dfrac{1}{2}E_1A\left( \dfrac{\pi}{l}\beta_1\cos\left(\dfrac{\pi x}{l}\right) + \dfrac{2\pi}{l}\beta_2\cos\left(\dfrac{2\pi x}{l}\right) \right)^2\\
    & \phantom{---} + \dfrac{1}{4}E_2A\left( \dfrac{\pi}{l}\beta_1\cos\left(\dfrac{\pi x}{l}\right) + \dfrac{2\pi}{l}\beta_2\cos\left(\dfrac{2\pi x}{l}\right) \right)^4\\
    & \phantom{---} + \alpha A \left( \dfrac{\pi}{l}\beta_1\cos\left(\dfrac{\pi x}{l}\right) + \dfrac{2\pi}{l}\beta_2\cos\left(\dfrac{2\pi x}{l}\right) \right)\left( \dfrac{\pi}{l}\dot{\beta_1}\cos\left(\dfrac{\pi x}{l}\right) + \dfrac{2\pi}{l}\dot{\beta_2}\cos\left(\dfrac{2\pi x}{l}\right) \right)
\end{align*}
The residual functions become
\begin{align*}
    \tilde{r} &= \rho A \ddot{\tilde{v}} - \tilde{\gamma}_x \tilde{v}_x - \tilde{\gamma} \tilde{v}_{xx} 
\end{align*}
and since we know that the inner product between the basis function and the residual function are equal due to their perpendicular traits
\begin{alignat*}{3}
    < \tilde{r},\phi_2>_{l}  &= \int_0^l{\tilde{r} \phi_1}{dx} &&= 0\\
    < \tilde{r}, \phi_2>_{l}  &= \int_0^l{\tilde{r} \phi_2}{dx} &&= 0
\end{alignat*}
This gives us the final output of two nonlinear ODEs
\begin{align*}
    &\dfrac{\rho Al}{2}\ddot{\beta}_1  + \dfrac{\alpha A}{l^3}\left(36.5284\beta_1^2 + 97.4091\beta^2_2\right)\dot{\beta}_1 + \dfrac{194.8182\alpha A}{l^3}\beta_1\beta_2\dot{\beta}_2 \\
    & \phantom{---} \dfrac{4.9348 T_0}{l}\beta_1 + \dfrac{E_1 A}{l^3}\beta_1\left(18.2642\beta_1^2 + 146.1136\beta_2^2\right) \\
    & \phantom{---} \dfrac{E_2A}{l^5}\beta_1\left( 75.1085\beta_1^4 + 2.1030\times 10^{3}\beta_1^2\beta_2^2 + 3.6052\times 10^{3}\beta_2^4\right) = 0\\[0.8cm]
    & \dfrac{\rho Al}{2}\ddot{\beta}_2  + \dfrac{\alpha A}{l^3}\left(97.4091\beta_1^2 + 584.4545\beta^2_2\right)\dot{\beta}_2 + \dfrac{194.8182\alpha A}{l^3}\beta_1\beta_2\dot{\beta}_1 \\
    & \phantom{---} \dfrac{19.7392 T_0}{l}\beta_1 + \dfrac{E_1 A}{l^3}\beta_2\left(146.1136\beta_1^2 + 292.2273\beta_2^2\right) \\
    & \phantom{---} \dfrac{E_2A}{l^5}\beta_2\left(1.0515\times 10^{3}\beta_1^4 + 7.2104\times 10^{3}\beta_1^2\beta_2^2 + 4.8069\times 10^{3}\beta_2^4\right) = 0
\end{align*}

## Simulations 

Based on the derivations above we can plot the phase plane of the two modes 

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp, DOP853
```

```python
# Defining parameters
params = {
    'rho': 500,
    'A': 0.0001,
    'l': 10,
    'T0': 1500,
    'E1': 2.5e+6,
    'E2': 0.5e+6,
    'alpha': 1.2e+6
}
```

```python
# System ODE
# x0 : beta1
# x1 : beta1_dot
# x2 : beta2
# x3 : beta2_dot
def nonlinear_string_ode(t, x, rho, A, l, T0, E1, E2, alpha):
    c11 = alpha*A/l**3 * (36.5284*x[0]**2 + 97.4091*x[2]**2)*x[1] 
    c12 = 194.8182*alpha*A/l**3 * x[0] * x[2] * x[3]
    c13 = 4.9348*T0/l * x[0]
    c14 = E1*A/l**3 * x[0] * (18.2642 * x[0]**2 + 146.1136 * x[2]**2)
    c15 = E2*A/l**5 * x[0] * (75.1085 * x[0]**4 + 2.1030e+3 * x[0]**2 * x[2]**2 + 3.6052e+3 * x[2]**4)
    beta1ddot = -2/rho/A/l * (c11 + c12 + c13 + c14 + c15)
    
    c21 = alpha*A/l**3 * (97.4091*x[0]**2 + 584.4545*x[2]**2)*x[3] 
    c22 = 194.8182*alpha*A/l**3 * x[0] * x[2] * x[1]
    c23 = 19.7392*T0/l * x[0]
    c24 = E1*A/l**3 * x[2] * (146.1136 * x[0]**2 + 292.2273 * x[2]**2)
    c25 = E2*A/l**5 * x[2] * (1.0515e+3 * x[0]**4 + 7.2104e+3 * x[0]**2 * x[2]**2 + 4.8069e+3 * x[2]**4)
    beta2ddot = -2/rho/A/l * (c21 + c22 + c23 + c24 + c25)
    
    return [x[1], beta1ddot, x[3], beta2ddot]
    
    
def solve_diffeq(func, t, tspan, ic, parameters={}, algorithm='DOP853', stepsize=np.inf):
    return solve_ivp(fun=func, t_span=tspan, t_eval=t, y0=ic, method=algorithm, 
                     args=tuple(parameters.values()), atol=1e-8, rtol=1e-5, max_step=stepsize)


def phasePlane(x1, x2, x3, x4, func, parameters):
    X1, X2 = np.meshgrid(x1, x2)  # create grid
    X3, X4 = np.meshgrid(x3, x4)  # create grid
    
    u1, v1 = np.zeros(X1.shape), np.zeros(X2.shape)
    u2, v2 = np.zeros(X3.shape), np.zeros(X4.shape)
    
    NI, NJ = X1.shape
    for i in range(NI):
        for j in range(NJ):
            x = X1[i, j]
            y = X2[i, j]
            xx = X3[i, j]
            yy = X4[i, j]
            dx = func(0, (x, y, xx, yy), *parameters.values())  # compute values on grid
            u1[i, j] = dx[0]
            v1[i, j] = dx[1]
            
            u2[i, j] = dx[2]
            v2[i, j] = dx[3]
    M1 = np.hypot(u1, v1)
    u1 /= M1
    v1 /= M1
    
    M2 = np.hypot(u2, v2)
    u2 /= M2
    v2 /= M2
    return X1, X2, u1, v1, M1, u2, v2, M2
```

```python
# x0 = np.random.uniform(-20, 20, (1,4))
x0 = [
    [-9.75191938, -2.91119727, -7.59715504, -2.7920785 ],
    [8.03450823, -17.2739244,   -5.20571273,  -6.70185079],
    [-17.37872695,  18.30880768,   1.79869491,   6.2961463 ],
    [ 6.72887358,  2.64431655, -7.31783452,  3.42364048]
]

x1, x2 = np.meshgrid(np.linspace(-20, 20, 20), np.linspace(-170, 170, 20))
x3, x4 = np.meshgrid(np.linspace(-30, 30, 20), np.linspace(-100, 100, 20))

# Set up the figure the way we want it to look
fig, ax = plt.subplots(2, 2, figsize=(11, 9), constrained_layout=True)

# Phase plane
b1, b1d, b2, b2d = nonlinear_string_ode(2, [x1, x2, x3, x4], *params.values())

m1 = np.hypot(b1, b1d)
m2 = np.hypot(b2, b2d)

i = 0
t1 = np.linspace(0, 200, 100000)
t_span1 = (np.min(t1), np.max(t1))
for x0i in x0:
    ax[i//2, i%2].quiver(x1, x2, b1/m1, b1d/m1, m1, scale=None, pivot='mid')
    sol1 = solve_diffeq(nonlinear_string_ode, t1, t_span1, x0i, params)
    ax[i//2, i%2].plot(sol1.y[0, :], sol1.y[1, :], '-')
    ax[i//2, i%2].plot(0, 0, '.r', ms=15)
    ax[i//2, i%2].set_title(r'$x_0$='+str(x0i))
    i+=1

fig.supxlabel(r'$\beta_1$')
fig.supylabel(r'$\dot{\beta}_1$')
plt.savefig('../plots/nl_string_mode1_phase.png')
plt.show()    
```

```python
fig, ax = plt.subplots(2, 2, figsize=(11, 9), constrained_layout=True)
i = 0
t1 = np.linspace(0, 200, 100000)
t_span1 = (np.min(t1), np.max(t1))
for x0i in x0:
    ax[i//2, i%2].quiver(x3, x4, b2/m2, b2d/m2, m2, scale=None, pivot='mid')
    sol1 = solve_diffeq(nonlinear_string_ode, t1, t_span1, x0i, params)
    ax[i//2, i%2].plot(sol1.y[2, :], sol1.y[3, :], '-')
    ax[i//2, i%2].plot(0, 0, '.r', ms=15)
    ax[i//2, i%2].set_title(r'$x_0$='+str(x0i))
    i+=1

fig.supxlabel(r'$\beta_2$')
fig.supylabel(r'$\dot{\beta}_2$')
plt.savefig('../plots/nl_string_mode2_phase.png')
plt.show()    
```

These demonstrate that the Kelvin-Voigt damping is allowing the system to damp to 0.
